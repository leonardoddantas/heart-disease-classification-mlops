# 📊 Exploratory Data Analysis (EDA)

This notebook presents the exploratory data analysis (EDA) of the heart disease dataset.  
The objective is to understand the data structure, identify potential issues, and support preprocessing decisions.

---

## 🧾 1. Import Libraries

**Description:**  
Import all necessary libraries for data manipulation and visualization.

---

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 📂 2. Load Dataset

**Description:**  
Load the dataset and inspect the first rows to understand its structure.

---

```python
df_heart_dissease = pd.read_csv("../data/raw/heart.csv")
df_heart_dissease.head()
```

## 🔍 3. Dataset Overview

**Description:**  
Analyze the dataset structure, including data types, number of entries, and memory usage.

---

```python
df_heart_dissease.shape
df_heart_dissease.info()
```

## 📊 4. Statistical Summary

**Description:**  
Generate summary statistics to understand the distribution of numerical features.

---

```python
df_heart_dissease.describe()
```

## ❌ 5. Missing Values Check

**Description:**  
Verify whether the dataset contains missing values.

---

```python
print(df_heart_dissease.isnull().sum())
```

## 🔁 6. Duplicate Rows

**Description:**  
Identify duplicated records that may affect model performance.

---

```python
print(df_heart_dissease.duplicated().sum())
```

## 📈 7. Distribution of Numerical Features

**Description:**  
Visualize the distribution of numerical variables to detect skewness and unusual patterns.

---

```python
numerical_cols = ["age", "resting_bp", "cholestoral", "max_hr", "oldpeak"]

plt.figure(figsize=(12, 8))

for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df_heart_dissease[col], kde=True)
    plt.title(col)

plt.tight_layout()
plt.show()
```

## 📦 8. Outlier Detection (Boxplot)

**Description:**  
Use boxplots to identify potential outliers in numerical features.

---

```python
sns.boxplot(data=df_heart_dissease)
plt.xticks(rotation=90)
plt.title("Weights of the variables")
plt.show()
```

## 📊 9. Categorical Variables Analysis

**Description:**  
Analyze the distribution and proportion of categorical variables.

---

```python
df_heart_dissease.hist(figsize=(12, 10))
plt.show()
```

## 🔥 10. Correlation Analysis

**Description:**  
Evaluate relationships between variables using a correlation matrix.

---

```python
plt.figure(figsize=(10,8))
sns.heatmap(df_heart_dissease.corr(), annot=True, cmap="coolwarm")
plt.show()
```

## 🎯 11. Target Variable Distribution

**Description:**  
Analyze the balance of the target variable to understand class distribution.

---

```python
sns.countplot(x="target", data=df_heart_dissease)
plt.show()
```

## ✅ Conclusion

The exploratory analysis helps identify key characteristics of the dataset, including:

- Absence or presence of missing values  
- Duplicate records  
- Distribution patterns and skewness  
- Presence of outliers  
- Balance of the target variable  
- Relationships between features  

These insights are essential for guiding the data preprocessing and modeling steps.
